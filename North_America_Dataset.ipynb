```python
import numpy as np
```

```python
total_rows = len(north_america_df)
```

```python
us_count = int(total_rows * 0.75)
canada_count = int(total_rows * 0.13)
mexico_count = total_rows - (us_count + canada_count)
```

```python
new_regions = (['United States'] * us_count) + (['Canada'] * canada_count) + (['Mexico'] * mexico_count)
```

```python
np.random.shuffle(new_regions)
```

```python
north_america_df.loc[north_america_df['Region_Category'] == 'North America', 'Region_Category'] = new_regions
```
