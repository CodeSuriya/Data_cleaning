```python
import pandas as pd
import os
```

```python
# Load the large dataset (Modify the filename accordingly)
file_path = "global_dataset.csv"  # Change to your actual dataset path
df = pd.read_csv("C:/Users/jsuri/Downloads/H_A/cleaned_global_dataset.csv")


```

```python
# Approximate file size in MB
total_size_mb = os.path.getsize('C:/Users/jsuri/Downloads/H_A/cleaned_global_dataset.csv') / (1024 * 1024)  # Convert bytes to MB
max_chunk_size = 100  # Maximum size of each split file in MB

# Calculate number of splits required
num_splits = int(total_size_mb // max_chunk_size) + 1

# Determine rows per split
rows_per_split = len(df) // num_splits

# Split and save files
for i in range(num_splits):
    start_row = i * rows_per_split
    end_row = start_row + rows_per_split if i != num_splits - 1 else len(df)
    
    df_chunk = df.iloc[start_row:end_row]
    
    # Save each chunk as a separate CSV file
    chunk_filename = f"global_dataset_part_{i+1}.csv"
    df_chunk.to_csv(chunk_filename, index=False)
    
    print(f"Saved: {chunk_filename} ({round(os.path.getsize(chunk_filename) / (1024 * 1024), 2)} MB)")

print("Dataset split successfully!")
```
